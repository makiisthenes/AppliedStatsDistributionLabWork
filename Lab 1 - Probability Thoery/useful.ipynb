```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    if x >= 0 and x < 1:
        return x
    elif x >= 1 and x <= 2:
        return 2 - x
    else:
        return 0

# Generate x values
x_values = np.linspace(-1, 3, 400)  # Adjust the range as needed

# Calculate corresponding y values
y_values = np.array([f(x) for x in x_values])

# Plot the function
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x)')
plt.grid(True)
plt.show()
```
